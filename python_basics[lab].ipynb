***11. Class***

Falls under a common paradigm OOP

Human:
properties- Name, Gender, Occupation
methods- speaks, work, sleep

*class is an abstraction of common set of properties and methods.*

**Object?**

"Albert Einstien and Aishwarya"

Instances of these two are different ex: name, gender, occupation etc.

````python
# n, o are arguments to the function
# name and occupation are properties of the class
# __init__ - initialize the properties of that particular class when later instance is created

class Human:
  def __init__(self, n, o):
    self.name = n
    self.occupation = o

  def work(self):
    if self.occupation == "Professor":
      print(self.name, 'teaches')
    elif self.occupation == "Actress":
      print(self.name, 'dance')

  def speaks(self):
    print(self.name, "says hello")

alb = Human("Albert", "Professor")
alb.work()
alb.speaks()

ai = Human("Aish", "Actress")
ai.work()
ai.speaks()
````
