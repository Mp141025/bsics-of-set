```python
import pandas as pd
import matplotlib.pyplot as plt  

```

```python
df = pd.read_csv("Downloads\company_sales_data.csv")

```

```python
profit = df ['total_profit'].tolist()
month  = df ['month_number'].tolist()
```

```python
plt.plot(month, profit, label = 'Line Plot of Total Profit all Months')
plt.xlabel('Month number')
plt.ylabel('TotAal Profit')
plt.xticks(month)
plt.title('Monthly Profit of the company')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```
